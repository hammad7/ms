```python
import logging

import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel
```

```python
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
train_df = pd.read_csv("./train.csv")
eval_df = pd.read_csv("./val.csv")
```

```python
train_df.rename(columns={"source":"input_text","target":"target_text"}, inplace=True)
eval_df.rename(columns={"source":"input_text","target":"target_text"}, inplace=True)
train_df.drop('Unnamed: 0', axis=1, inplace=True)
eval_df.drop('Unnamed: 0', axis=1, inplace=True)
eval_df
```

```python
MAX_LENGTH = 300
MAX_SEQ_LENGTH = 100
```

```python
# train_data = [
#     ["one", "1"],
#     ["two", "2"],
# ]

# train_df = pd.DataFrame(train_data, columns=["input_text", "target_text"])

# eval_data = [
#     ["three", "3"],
#     ["four", "4"],
# ]

# eval_df = pd.DataFrame(eval_data, columns=["input_text", "target_text"])

model_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "max_seq_length": MAX_SEQ_LENGTH,
    "train_batch_size": 32, ##
    "num_train_epochs": 20, ## 200
    "save_eval_checkpoints": False,
    "save_model_every_epoch": False,
    "evaluate_generated_text": True,
    "evaluate_during_training": False,
    "evaluate_during_training_verbose": True,
    "use_multiprocessing": False,
    "repetition_penalty": 1, ###
    "num_beams": 1,
    "max_length": MAX_LENGTH,
    "manual_seed": 4,
}

encoder_type = "roberta"

model = Seq2SeqModel(
    encoder_type,
    "roberta-base",
    "bert-base-cased",
    args=model_args,
    use_cuda=True, #### True
)

```

```python
print(train_df.loc[0:10])
print(eval_df.loc[0:10])
```

```python
model.train_model(train_df.loc[0:5000])
```

```python
results = model.eval_model(eval_df.loc[0:100])
print(results)
```
